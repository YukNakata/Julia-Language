## Ikeda eq Delay Differential Equations

```julia
using DifferentialEquations
using QuadGK
using Plots
#using Interact
using Measures
#using LaTeXStrings
gr()
```

```julia
L"\tau"
```

```julia


function ikedadde(du,u,h,p,t)
du[1] = -sin(h(p,t-p[1])[1])
end

# Initial function


h(p, t) = 0.5*ones(1)
u0 = [0.5]
## Initial function

p=[1.0]
lags = [p[1]]
tspan = (0.0,80.0)

prob = DDEProblem(ikedadde,u0,h,tspan,p; constant_lags=lags)

#sol = solve(prob)
alg = MethodOfSteps(Tsit5())
sol = solve(prob,alg,reltol=1e-6)
```

```julia
plot(sol,
    vars=(0,1),
    linewidth=4,
    color="#6a656d",
    label="x",
    xtickfont = font(16),
#    xticks = 0:2:200,
    ytickfont = font(16)) # legend=false
x = range(-1,stop=0,length=10)
plot!(x,x->1.0,
    linewidth=4,
    color="#FF9900",label="x init")
plot!(xlim=(-1,80))
```

```julia
animw=@animate for i=0:6
    
p=[i*0.5,1.0]

prob = DDEProblem(ikedadde,u0,h,tspan,p; constant_lags=lags)

sol = solve(prob,alg,reltol=1e-6)
    
plot(sol,
    vars=(0,1),
    linewidth=4,
    color="#6a656d",
    xtickfont = font(16),
    legendtitle="tau=$(i*0.5)",
    label="",
    ytickfont = font(16),
    legendfontsize=12) # legend=false
x = range(-i*0.25,stop=0,length=10)
plot!(x,x->0.5,
    linewidth=4,
    color="#FF9900",
        label=""
    )
plot!(xlim=(-1,100))
end
```

```julia
tspan = (0.0,80.0)
p=[0.5,1.0]

prob = DDEProblem(ikedadde,u0,h,tspan,p; constant_lags=lags)

sol1 = solve(prob,alg,reltol=1e-6)

p=[1.5,1.0]

prob = DDEProblem(ikedadde,u0,h,tspan,p; constant_lags=lags)

sol2= solve(prob,alg,reltol=1e-6)

p=[2.5,1.0]

prob = DDEProblem(ikedadde,u0,h,tspan,p; constant_lags=lags)

sol3= solve(prob,alg,reltol=1e-6)

p=[3.5,1.0]

prob = DDEProblem(ikedadde,u0,h,tspan,p; constant_lags=lags)

sol4= solve(prob,alg,reltol=1e-6)

p=[4.5,1.0]

prob = DDEProblem(ikedadde,u0,h,tspan,p; constant_lags=lags)

sol5= solve(prob,alg,reltol=1e-6)
```

```julia
l = @layout [a;b ; c ;d;e]
p2 = plot(sol1,
    color="#6a656d",
    vars=(0,1),
    linewidth=4,
    yticks = 0:0.5:1.0,
    xtickfont = font(16),
    label="",
    ytickfont = font(16),
    legendfontsize=12,
        bottom_margin=3mm,
    right_margin=3mm,
        legendtitle="tau=0.5") # legend=false
p3 = plot(sol2,
        color="#6a656d",
    vars=(0,1),
        yticks = -0.5:0.5:1.5,
    linewidth=4,
    xtickfont = font(16),
    label="",
    ytickfont = font(16),
    legendfontsize=12,
            legendtitle="tau=1.5") # legend=false
p4 = plot(sol3,
        color="#6a656d",
    vars=(0,1),
        yticks = -3:1.0:3.0,
    linewidth=4,
    xtickfont = font(16),
    label="",
    ytickfont = font(16),
    legendfontsize=12,
                legendtitle="tau=2.5") # legend=false

p5 = plot(sol4,
        color="#6a656d",
        yticks = -4:2.0:4.0,
    vars=(0,1),
    linewidth=4,
    xtickfont = font(16),
    label="",
    ytickfont = font(16),
    legendfontsize=12,
    legendtitle="tau=3.5",
    legend=:bottomright
)
p6 = plot(sol5,
        color="#6a656d",
        yticks = -4:2.0:4.0,
    vars=(0,1),
    linewidth=4,
    xtickfont = font(16),
    label="",
    ytickfont = font(16),
    legendfontsize=12,
    legendtitle="tau=4.5",
    legend=:bottomright
)
plot(p2, p3, p4, p5,p6, size = (900, 1000), layout = l)
```
