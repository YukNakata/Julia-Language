```julia
using Plots
theme(:ggplot2)
```

```julia
p0=1/6
p1=2/6
p2=3/6
```

```julia
R0=1/6*0+2/6*1+3/6*2
```

```julia
function brprocess(x)
x_new=zeros(x)
a=rand(x) #乱数を感染者の数だけ発生

for i=1:x　#i番目の感染者が再生産する感染者数を決定
    if a[i]<p0
    x_new[i]=0
    elseif a[i]<p0+p1
    x_new[i]=1
    else
    x_new[i]=2            
    end
end

x2=0　#次世代の感染者数を計算
for i=1:x
x2=x_new[i]+x2
end
x=x2
return x
end
```

```julia
gen=20
x = append!([1],zeros(gen))
for i in 1:gen
    x[i+1]= brprocess(x[i]) 
end
#print(x)
plot(x,
    line=2,
    color="#F37121",
    size=(600,300),
    marker=:circle,
    markercolor = "#F37121",
    markerstrokecolor="#F37121",
    legend=:false) 
```

```julia
gen=15
sim=9

xlist=ones(Int64, gen+1,sim)

for j in 1:sim
    x=1
    x_seq=[x]
for i in 1:gen
    x= brprocess(x)
    x=Int(x)
    x_seq=vcat(x_seq,x)
end
    xlist[:,j]=x_seq
end

plot(xlist, 
    layout = (3,3),
    size=(700,400),
    legend=:false,
    line=2,
    color="#F37121",
    marker=:circle,
    markersize  = 3,
    markercolor = "#F37121",
    markerstrokecolor="#F37121")
```
