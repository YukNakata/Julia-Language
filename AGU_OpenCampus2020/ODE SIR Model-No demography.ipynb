# ODE SIR Model (No demography)

```julia
using DifferentialEquations
using QuadGK
using Plots
theme(:ggplot2)
using Measures
```

```julia
function odesir(du,u,p,t)
  du[1] = -p[1]*u[1]*u[2]
  du[2] = p[1]*u[1]*u[2]-p[2]*u[2]
end

# Parameter 
p=[[2.0*1/10,1/10],[1.5*1/10,1/10],[0.8*1/10,1/10]]
```

```julia
tspan = (0.0,1000.0)
```

```julia
u0 = [0.999,0.001]

prob1 =ODEProblem(odesir,u0,tspan,p[1]);
prob2 =ODEProblem(odesir,u0,tspan,p[2]);
prob3 =ODEProblem(odesir,u0,tspan,p[3]);

sol1=solve(prob1)
sol2=solve(prob2)
sol3=solve(prob3)
```

```julia
plot(sol1,
    vars=(0,1),
    linewidth=4,
    color="#21a3f3",
    label=""
    ) # legend=false

plot!(sol1,
    vars=(0,2),
    linewidth=4,
    color="#F37121",
    label=""
    ) # legend=false

plot!(
#    xtickfont = font(14),
    xlim=(0,250),
#    ytickfont = font(14),
   size=(600,350),
#    bottom_margin=3mm,
#    right_margin=2mm,
#    background_color="#e6e6fa"
    ) # legend=false
```
